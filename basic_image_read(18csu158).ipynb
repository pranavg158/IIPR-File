```python
#to install skimage  use pip install scikit-image
from skimage import io,img_as_float
img=io.imread("cricket.jpg",as_gray=True)
```

```python
print(img.shape) #y,x,c
```

```python
print(img)
```

```python
print(io.imshow(img))
```

```python
img2=img_as_float(img)
```

```python
print(img2)
```

```python
from skimage import img_as_ubyte
img_8bit=img_as_ubyte(img2)
```

```python
print(img_8bit)
```

```python
print(io.imshow(img_8bit))
```

```python
#filtering
from skimage import filters
guassian_img=filters.gaussian(img,sigma=1)
print(guassian_img)
```

```python
io.imsave("saved_image.jpg",guassian_img)
```

```python
from skimage import img_as_ubyte
guassian_img_8bit=img_as_ubyte(guassian_img)
io.imsave("saveed_image_8bit.tif",guassian_img_8bit)
```

```python
# load and show an image with Pillow
from PIL import Image
from numpy import asarray
# Open the image form working directory
image = Image.open('cricket.jpg')
# summarize some details about the image
print(image.format)
print(image.size)
print(image.mode)
# show the image
image.show()

# convert image to numpy array
data = asarray(image)
print(type(data))
# summarize shape
print(data.shape)


# create Pillow image
image2 = Image.fromarray(data)
print(type(image2))

# summarize image details
print(image2.mode)
print(image2.size)
```

```python
print(data)
```

```python
#Converting Images to Grayscale
import numpy as np
im = np.array(Image.open('cricket.jpg').convert('L')) #you can pass multiple arguments in single line
print(im.shape)
print(type(im))

gr_im= Image.fromarray(im).save('cricket.jpg')
print(gr_im)
```

```python
#Resizing 
load_img_rz = np.array(Image.open('cricket.jpg').resize((200,200)))
Image.fromarray(load_img_rz).save('cricket.jpg')
print("After resizing:",load_img_rz.shape)
```

```python
#trimming
im = np.array(Image.open('ball.jpg'))

print("Before trimming:",im.shape)

im_trim = im[128:384, 128:384]
print("After trimming:",im_trim.shape)

Image.fromarray(im_trim).save('trim_saltandpepper.jpg')
```

```python
# load and display an image with Matplotlib
from matplotlib import image
from matplotlib import pyplot
# load image as pixel array
image = image.imread('ball.jpg')
print(image)
# summarize shape of the pixel array
print(image.dtype)
print(image.shape)
# display the array of pixels as an image
pyplot.imshow(image)
pyplot.show()
```
